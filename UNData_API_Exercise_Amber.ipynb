Q1

```python
import requests
import pandas as pd

url = "http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD"
params = {"format": "json", "per_page": 20000}

respones = requests.get(url, params=params)
data = respones.json()

gdp_df = pd.json_normalize(data[1])
gdp_df.head()
```

Q2.

```python
url = "http://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN"
params = {"format": "json", "per_page": 20000}

respones = requests.get(url, params=params)
data = respones.json()

life_exp_df = pd.json_normalize(data[1])
life_exp_df.head()
```

Q3.

```python
gdp_clean = gdp_df[["countryiso3code", "country.value", "date", "value"]].rename(
    columns={"value": "gdp_per_capita_ppp"})
    
Life_clean = life_exp_df[["countryiso3code", "country.value", "date", "value"]].rename(
    columns={"value": "life_expectancy"})

merged_df = pd.merge(  gdp_clean, Life_clean,
    on=["countryiso3code", "country.value", "date"],
    how="outer")

merged_df.head()            
```

Q4.

```python
url = "http://api.worldbank.org/v2/country"
params = {"format": "json", "per_page": 400}

respones = requests.get(url, params=params)
data = respones.json()

country_df = pd.json_normalize(data[1])
country_df.head()
```

```python
country_clean = country_df[["id", "region.value", "capitalCity", "incomeLevel.value"]].rename(
    columns={"id": "countryiso3code"})

full_df = pd.merge(merged_df, country_clean, on="countryiso3code", how="left")

full_df = full_df[full_df["region.value"] != "Aggregates"]

full_df.head()

    
```
